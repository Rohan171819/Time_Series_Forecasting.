```python
 import pandas as pd
 import numpy as np
```

```python
data=pd.read_csv("/content/Tesla.csv - Tesla.csv.csv")
```

```python
data.head()
```

# Simple Moving Average

```python
data["Open"].plot(ls="--",color="black",figsize=(12,5))
```

```python
data["Open: 10 Days Rolling"]=data[["Open"]].rolling(window=10,min_periods=1).mean()
data["Open: 30 Days Rolling"]=data[["Open"]].rolling(window=30,min_periods=1).mean()
data["Open: 50 Days Rolling"]=data[["Open"]].rolling(window=50,min_periods=1).mean()
data["Open: 70 Days Rolling"]=data[["Open"]].rolling(window=70,min_periods=1).mean()
data["Open: 90 Days Rolling"]=data[["Open"]].rolling(window=90,min_periods=1).mean()
```

```python
data[["Open","Open: 10 Days Rolling","Open: 30 Days Rolling","Open: 50 Days Rolling","Open: 70 Days Rolling","Open: 90 Days Rolling"]].plot(ls="--",figsize=(15,6))
```

# Commulative Moving Average

```python
# Expanding Method...
data[["Open"]].expanding().mean().plot(ls="--",color="red")
```

# Exponential Moving Average

```python
data["Open: EWA_0.1"]=data[["Open"]].ewm(alpha=0.1,adjust=False).mean()
data["Open: EWA_0.3"]=data[["Open"]].ewm(alpha=0.3,adjust=False).mean()
data["Open: EWA_0.5"]=data[["Open"]].ewm(alpha=0.5,adjust=False).mean()
data["Open: EWA_0.7"]=data[["Open"]].ewm(alpha=0.7,adjust=False).mean()
data["Open: EWA_0.9"]=data[["Open"]].ewm(alpha=0.9,adjust=False).mean()
```

```python
data[["Open","Open: EWA_0.1","Open: EWA_0.3","Open: EWA_0.5","Open: EWA_0.7","Open: EWA_0.9"]].plot(ls="--",figsize=(15,6))
```
