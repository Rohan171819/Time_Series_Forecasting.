```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.api import ExponentialSmoothing,SimpleExpSmoothing,Holt
```

```python
Data=pd.read_csv("/content/rainfall in india 1901-2015.csv")
```

```python
# Visualizing the Dataset for Jan-Feb Months..
plt.plot(Data[1:100]["Jan-Feb"])
plt.show()
```

```python
# Visualizing the Dataset for Mar-May Months..
plt.plot(Data[1:100]["Mar-May"])
plt.show()
```

```python
# Visualizing the Dataset for Jun-Sep Months..
plt.plot(Data[1:100]["Jun-Sep"])
plt.show()
```

```python
# Visualizing the Dataset for Oct-Dec Months..
plt.plot(Data[1:100]["Oct-Dec"])
plt.show()
```

# Moving Average...

```python
# Rolling Average Function...
Rolling_Series_Data=Data[1:100]["Jun-Sep"].rolling(window=20)
Rolling_Series_Average=Rolling_Series_Data.mean()
```

```python
Rolling_Series_Average.head(24)
```

```python
plt.plot(Rolling_Series_Average,color="blue")
plt.show()
```

# Simple Exponential Smoothing...

```python
data=Data[1:100]["Jun-Sep"]
fit1=SimpleExpSmoothing(data).fit(smoothing_level=0.9,optimized=False)
fit2=SimpleExpSmoothing(data).fit(smoothing_level=0.2,optimized=False)
```

```python
plt.plot(data,color="black",marker="*")
plt.plot(fit1.fittedvalues,color="red",marker="*")
plt.plot(fit2.fittedvalues,color="blue",marker="*")
plt.show()
```

# Holt Function...

```python
fit1=Holt(data,exponential=False).fit()  # Assunig that there is Linear Trend in the data...
#fit2=Holt(data,exponential=True).fit()  # Assuming that there is Exponential Trend in the data...
```

```python
plt.plot(data,color="black",marker="*")
plt.plot(fit1.fittedvalues,color="red",marker="*")
plt.show()
```
